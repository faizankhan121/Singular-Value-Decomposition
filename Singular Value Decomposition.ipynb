```python
#Read Dataset
read=pd.read_csv("temp_data.csv")
read
```

```python
#Load Dataset as Matrix
matrix=read.to_numpy()
matrix
```

```python
#find transpose of given matrix
transpose = [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
print("Transpose is:- ")
for transposed in transpose:
    print(transposed)
```

```python
square =  [[sum(first*second for first,second in zip(X_rows,Y_column)) for Y_column in zip(*transpose)] for X_rows in matrix]
print("Square Matrix is:- ")
for dot in square:
    print(dot)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(square)
idx = eigenvalues.argsort()[::-1]   
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:,idx]
```
